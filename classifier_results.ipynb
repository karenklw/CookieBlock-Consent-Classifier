Creating Dataframes with Columns for Use in Getting Ratios for CDFs

```python
def create_dataframe_from_json(file_path, mode):
    with open(file_path, 'r') as json_file:
        json_data = json.load(json_file)

    count_necessary = {} # mapped to a 0 in JSON file
    count_functional = {} # mapped to a 1 in JSON file
    count_analytics = {} # mapped to a 2 in JSON file
    count_advertising = {} # mapped to a 3 in JSON file
    site_urls = set()

    for key, value in json_data.items():
        site_url = key.split(';')[2] #FIXME: remove the "/" before URL
        site_urls.add(site_url)

        if value == 0:
            count_necessary[site_url] = count_necessary.get(site_url, 0) + 1
        elif value == 1:
            count_functional[site_url] = count_functional.get(site_url, 0) + 1
        elif value == 2:
            count_analytics[site_url] = count_analytics.get(site_url, 0) + 1
        elif value == 3:
            count_advertising[site_url] = count_advertising.get(site_url, 0) + 1

    # lists for DataFrame columns
    site_urls = list(site_urls)
    interactions = [mode] * len(site_urls)
    necessary_counts = [count_necessary.get(site_url, 0) for site_url in site_urls]
    functional_counts = [count_functional.get(site_url, 0) for site_url in site_urls]
    analytics_counts = [count_analytics.get(site_url, 0) for site_url in site_urls]
    advertising_counts = [count_advertising.get(site_url, 0) for site_url in site_urls]

    data = {
        "site_url": site_urls,
        "interaction": interactions,
        "necessary": necessary_counts,
        "functional": functional_counts,
        "analytics": analytics_counts,
        "advertising": advertising_counts
    }
    df = pd.DataFrame(data)

    return df
```

CDFs for Distribution of Categories Across Websites

```python
def get_ratios(df_normal, df_reject):
    ratios = {}
    categories = ["necessary", "functional", "analytics", "advertising"]

    for category in categories:
        normal_counts = df_normal[category]
        reject_counts = df_reject[category]

        # handle cases where a site has data in df_normal but not in df_reject
        for site_url in df_normal["site_url"]:
            if site_url not in df_reject["site_url"]:
                reject_counts.loc[site_url] = 0

        ratios[category] = reject_counts / normal_counts

    return ratios


"""
Returns a dataframe which can be plotted as a CDF graph (see plot below).
`list` should be a list of values
"""
def get_cdf(list):
    s = pd.Series(list, name = 'value')
    df = pd.DataFrame(s)

    # Frequency
    stats_df = df \
    .groupby('value') \
    ['value'] \
    .agg('count') \
    .pipe(pd.DataFrame) \
    .rename(columns = {'value': 'frequency'})

    # PDF
    stats_df['pdf'] = stats_df['frequency'] / sum(stats_df['frequency'])

    # CDF
    stats_df['cdf'] = stats_df['pdf'].cumsum()
    stats_df = stats_df.reset_index()

    return stats_df
```

CDF of Request+Response Cookies Collected During [Reject Run] / [Normal Run], Classified Using 20231002 Model

```python
sites_to_category_counts_normal = create_dataframe_from_json("results/predictions_new_data/normal_predictions_nolabel.json", "normal")
# print(sites_to_category_counts_normal)

sites_to_category_counts_reject = create_dataframe_from_json("results/predictions_new_data/reject_predictions_nolabel.json", "reject")
# print(sites_to_category_counts_reject)

# sites = sites_to_category_counts_normal["site_url"].tolist()
# print(sites)

ratios = get_ratios(sites_to_category_counts_normal, sites_to_category_counts_reject)

# CDF plots for each category
for category, ratio in ratios.items():
    cdf_data = get_cdf(ratio)
    plt.plot(cdf_data["value"], cdf_data["cdf"], label=category)

# plt.axis([0.4, 2, 0, 1]) # comment/uncomment this to restrict size of plot
plt.title("CDF of Reject/Normal Ratios by Class, 20231002 Model")
plt.xlabel("Reject / Normal Ratio")
plt.ylabel("CDF")
plt.legend()
plt.grid(visible=True)
plt.gcf().set_facecolor('white')
plt.show()
```
